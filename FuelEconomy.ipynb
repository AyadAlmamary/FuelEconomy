```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
FuelEconomy = pd.read_csv('FuelEconomy.csv')

FuelEconomy.info()

```

```
sns.jointplot( x ='Horse Power',y = 'Fuel Economy (MPG)', data = FuelEconomy , color = 'red')

sns.pairplot(FuelEconomy)

sns.lmplot( x ='Horse Power',y = 'Fuel Economy (MPG)', data = FuelEconomy)

X = FuelEconomy[['Horse Power']]
y = FuelEconomy['Fuel Economy (MPG)']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)

print('m_coeff = ',regressor.coef_ )
print('b_coeff = ',regressor.intercept_ )

y_predict = regressor.predict(X_test)

plt.scatter(X_train, y_train, color= 'green')
plt.plot(X_train, regressor.predict(X_train), color= 'red')

plt.ylabel('Fuel Economy (MPG)')
plt.xlabel('Horse Power [HP]')
plt.title('Fuel Economy (MPG) vs. Horse Power')
```
